#How many content are released each year?

```python
release_count= data.Years.value_counts() #shows number of content released in each year
release_count
```

```python
plt.figure(figsize=(10,6))
release_count.plot(kind='bar', color='skyblue')
plt.title('Number of Content Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#Name the top 10 directors which has most number of releases

```python
top_directors= data.Director.value_counts().head(10).sort_values(ascending=False)
top_directors
```

```python
plt.figure(figsize=(10,6))
top_directors.plot(kind='bar', color='royalblue')
plt.title('Top 10 directors with most content released')
plt.xlabel('Names')
plt.ylabel('Number of Releases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#Show me the category wise content count
